```python
import pandas as pd
import numpy as np
```

```python


# Veri çerçevesini yükle
loans_2007 = pd.read_csv('LoanStats3a.csv', skiprows=1, low_memory=False)

# Eksik değerleri olan sütunları kaldır
half_count = len(loans_2007) / 2
loans_2007 = loans_2007.dropna(thresh=half_count, axis=1)

# 'url' ve 'desc' sütunlarını kaldır
loans_2007 = loans_2007.drop(['url', 'desc'], axis=1)

# İlk 5 satırı ve sütunların bilgilerini göster
loans_2007.head()

```

```python
print('of rows and columns of loans_2007 = {} , {} '.format(loans_2007.shape[0],loans_2007.shape[1]))
```

## Veri temizliği

```python
removed_columns = ['id','member_id','funded_amnt','funded_amnt_inv','grade','sub_grade','emp_title',
                   'issue_d','zip_code','out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv',
                   'total_rec_prncp','total_rec_int','total_rec_late_fee',
                   'recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt']

loans_2007 =loans_2007.drop(removed_columns, axis=1)
```

```python
print('of rows and columns of loans_2007 = {} , {} '.format(loans_2007.shape[0],loans_2007.shape[1]))
```

```python
loans_2007.head()
```

##### Hedef değerimiz (loan_status)

```python
print(loans_2007.loan_status.value_counts())
```

```python
loans_2007 =loans_2007[(loans_2007['loan_status'] ==  'Fully Paid') | 
                       (loans_2007['loan_status'] ==  'Charged Off')]
```

İstediğimiz değerler yapacağımız algoritmaya göre olacağından işimize yarayan bilgiler kredi odeme bilgisi olduğu için geri kalan bilgileri silmiş olduk

```python
mapping_dict = {
    "loan_status" : {
        "Fully Paid" : 1,
        "Charged Off" :0
    }
}

loans_2007 =loans_2007.replace(mapping_dict)
```

Algorıtmalarda kullanın değerler sayılardır bundan dolayı object değerleri 1 ve 0 dönüştürmüş olduk 

```python
loans_2007.head(5)
```

```python
drop_columns = []
for col in loans_2007.columns:
    is_unique = len(loans_2007[col].dropna().unique())
    if is_unique == 1:
        drop_columns.append(col)
```

```python
drop_columns
```

```python
loans_2007 = loans_2007.drop(drop_columns, axis=1)
```

Veri setimizdeki columnlardan hep aynı değerleri taşıması bizim için anlam ifade etmeyeceği için bu columnları basit bir fonksiyonla belirledikten sonra silmiş olduk
