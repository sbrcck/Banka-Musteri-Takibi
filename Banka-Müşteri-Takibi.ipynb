### Tahmin Algoritması ile Kredi Modelleme

```python
print(loans_2007.info())
```

```python
features = loans_2007.drop('loan_status', axis=1)
```

```python
features.head()
```

```python
target = loans_2007['loan_status']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
lr = LogisticRegression(class_weight= 'balanced') 
```

```python
predictions= cross_val_predict(lr, features, target, cv=3)
```

```python
predictions= pd.Series(predictions)
```

```python
predictions.value_counts()
```

```python
fp_filter = (predictions == 1) & (target == 0)
fp = len(predictions[fp_filter])

tp_filter = (predictions == 1) & (target == 1)
tp = len(predictions[tp_filter])

fn_filter = (predictions == 0) & (target == 1)
fn = len(predictions[fn_filter])

tn_filter = (predictions == 0) & (target == 0)
tn = len(predictions[tn_filter])

tpr = tp / (tp + fn)
fpr = fp / (fp + tn)

print(tpr)
print(fpr)
```

```python
# Sınıf ağırlıklarını belirleyen "penalty" sözlüğü oluşturuluyor.
penalty = {
    0: 10,
    1: 1
}

# LogisticRegression sınıfından bir model oluşturuluyor ve sınıf ağırlıkları belirtiliyor.
#lr = LogisticRegression(class_weight=penalty)
lr = LogisticRegression(class_weight=penalty)


# Çapraz doğrulama yaparak tahminler elde ediliyor ve bir pandas Serisi olarak saklanıyor.
predictions = cross_val_predict(lr, features, target, cv=3)
predictions = pd.Series(predictions)

# Yanlış pozitifleri (False Positives) bulmak için filtre oluşturuluyor ve sayısı hesaplanıyor.
fp_filter = (predictions == 1) & (target == 0)
fp = len(predictions[fp_filter])

# Doğru pozitifleri (True Positives) bulmak için filtre oluşturuluyor ve sayısı hesaplanıyor.
tp_filter = (predictions == 1) & (target == 1)
tp = len(predictions[tp_filter])

# Yanlış negatifleri (False Negatives) bulmak için filtre oluşturuluyor ve sayısı hesaplanıyor.
fn_filter = (predictions == 0) & (target == 1)
fn = len(predictions[fn_filter])

# Doğru negatifleri (True Negatives) bulmak için filtre oluşturuluyor ve sayısı hesaplanıyor.
tn_filter = (predictions == 0) & (target == 0)
tn = len(predictions[tn_filter])
```

```python
# Duyarlılık (Sensitivity) hesaplanıyor.
tpr = tp / (tp + fn)

# Yanlış pozitif oranı (False Positive Rate) hesaplanıyor.
fpr = fp / (fp + tn)

# Duyarlılık ve yanlış pozitif oranı ekrana yazdırılıyor.
print(tpr)
print(fpr)
```
